```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = np.zeros((600,600,3), np.uint8)
cv2.line(img,(0,0),(600,600),(255,0,0),5)
cv2_imshow(img)
```

```python
img = np.zeros((600,600,3), np.uint8)
cv2.line(img,(0,0),(600,600),(255,0,0),5)
cv2.line(img,(600,0),(0,600),(20,255,0),10)
cv2_imshow(img)
```

```python
img = np.zeros((600,600,3), np.uint8)
cv2.line(img,(0,0),(600,600),(255,0,0),5)
cv2.line(img,(600,0),(0,600),(20,255,0),10)
cv2.circle(img,(300,300),10,(0,0,255),5)
cv2_imshow(img)
```

```python
img = np.zeros((600,600,3), np.uint8)
cv2.line(img,(0,0),(600,600),(255,0,0),5)
cv2.line(img,(600,0),(0,600),(20,255,0),10)
cv2.circle(img,(300,300),10,(0,0,255),5)
cv2.rectangle(img,(0,0),(100,150),(0,0,255),5)
cv2_imshow(img)
```

```python
img = np.zeros((600,600,3), np.uint8)
cv2.line(img,(0,0),(600,600),(255,0,0),5)
cv2.line(img,(600,0),(0,600),(20,255,0),10)
cv2.circle(img,(300,300),50,(0,0,255),2)
cv2.rectangle(img,(0,0),(100,150),(0,0,255),5)
cv2.rectangle(img,(350,350),(250,250),(255,255,0),2)
cv2_imshow(img)
```
